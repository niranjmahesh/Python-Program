```python
# Guess the number generated by computer

import random

end_range = int(input("Please provide the range"))
guess_count = int(input("please enter the number of guesses"))
start_range = 1
random_number = random.randint(start_range, end_range)

while guess_count  != 0:
             
        status_feedback  = str(input(f" Is {random_number} high (H) or Low (L) or Correct (C)"))
             
        if status_feedback == 'C' :
            print(f"Your Guessed number is {random_number} ")
            guess_count  = 1
        elif status_feedback == 'H':
            end_range = random_number
            random_number = random.randint(start_range, end_range)
            print(f'H {start_range} {end_range}')
        elif status_feedback == 'L':
             start_range = random_number
             random_number = random.randint(start_range, end_range)
             print(f'L {start_range} {end_range}')
             
        guess_count = guess_count - 1 

```
