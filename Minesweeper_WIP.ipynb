```python
import random
import os


class board :
    def __init__(self,dim_size,num_bombs):
        self.dim_size = dim_size     # board size
        self.num_bombs = num_bombs   # number of bombs
        self.board = self.make_new_board()
        self.assign_values_to_board() #assign the neigboring bombs 
        self.dug = set()                #dig status
      
  
        
    def make_new_board(self):
                
        board = [[None]*self.dim_size for _ in range(self.dim_size)]

        bombs_planted = 0 
        while bombs_planted < self.num_bombs:   # loop until the bombs planted is less than number of bombs
            loc = random.randint(0, self.dim_size**2 -1) #random number generation
            row = loc // self.dim_size                  #whole number of the division
            col = loc % self.dim_size                   #reminder of the division
            #print(f"{loc} {row} {col}")
            if board [row] [col] == '*': # bomb is already placed
                continue
            else:
                board [row] [col] = '*'
                bombs_planted +=1
             #   print(f"print bombs planted {bombs_planted}")
        return (board)
    
    def __str__(self):
        board_str = '   ' + ' '.join([str(i) for i in range(self.dim_size)]) + '\n'
        for row_idx in range(self.dim_size):
            row_str = f"{row_idx:2} "
            for col_idx in range(self.dim_size):
                if (row_idx, col_idx) in self.dug:
                    row_str += f"{self.board[row_idx][col_idx]}  "
                else:
                    row_str += ' ?  '
            board_str += row_str + '\n'
        return board_str
    
    '''
     /*   # Create a header row with column numbers
        board_str = ' '  # Starting offset for column labels
        #board_str = '    ' 
        for col in range(self.dim_size):
            board_str += f"{col:3} "  # Format each column index with 5 spaces for alignment
        board_str += '\n'
        
        # Create the board with row numbers and corresponding cells
        #for row_idx, row in enumerate(self.board):
        for row_idx, row in enumerate(self.board):
                row_str = f"{row_idx:2} "  # Add row number
                row_str += '   '.join([str(cell) for cell in row ]) # Add row cells
                #row_str += ' '.join([str(cell) if cell is not None else '|0|' for cell in row])  # Add row cells
                board_str += row_str + '\n'
          
       
        return board_str

'''

    def assign_values_to_board(self):
        
        for r in range(self.dim_size):
                for c in range(self.dim_size):
                     if self.board [r] [c] == '*':
                        continue
                     self.board [r] [c] = self.get_number_neighboring_bombs(r,c)

    def get_number_neighboring_bombs(self,row,col):
                
                number_neighboring_bombs = 0

                for r in range(max(0,row-1) , min(self.dim_size , row+2)):
                     for c in range(max(0,col-1) , min(self.dim_size, col+2)):
                          if r == row and c == col:
                            continue
                          if self.board [r] [c] =='*':
                            number_neighboring_bombs +=1
                return (number_neighboring_bombs)
                     

    def dig(self , row , col):
        if (row , col) in self.dug : # check  in the list
            return False 
        self.dug.add((row ,col))


        if self.board[row][col] == '*':
              return True # Its a bomb
        
         # If no bombs around, dig all neighboring cells recursively
        if self.board[row][col] == 0:
            for r in range(max(0, row - 1), min(self.dim_size, row + 2)):
                for c in range(max(0, col - 1), min(self.dim_size, col + 2)):
                    if (r, c) not in self.dug:
                        self.dig(r, c)  # Recursively dig neighboring cells
        return False

def clear():
    os.system('cls' if os.name == 'nt' else 'clear')
          

def play_game():
     
    x = int(input(f"Please enter the dimension"))
    y = int(input(f"Please enter the number of bombs"))

    if y >= x * x:
        print("The number of bombs cannot exceed the total number of cells on the board!")
        
    
    # Create a new board with the provided dimension and number of bombs
    new_board = board(x, y)
    

    # Print the generated board
    print(f"Here is your board:\n{new_board}")
        #print(f"Here is your board:\n{new_board.assign_values_to_board()}")

    game_over = False

    while not game_over :

        inp_row , inp_col  = map(int, input("Enter row number followed by space and column number = ").split())
        
    if new_board.dig(inp_row, inp_col):
            print("Boom ! You have hit a  bomb")
            game_over = True
    else:
            print("Safe ! No Bombs")
            print(new_board)


    if len(new_board.dug) == (x * x - y):  # All cells are dug except bombs
            print("Congratulations! You won!")
            game_over = True

if __name__ == '__main__':
    play_game()           
        






```
